```python
class TicTacToe():
  def __init__(self, map_grid_size, discount):

    # Constructor initliazes relevant parameters to the value iteration like the discount rate, number of iterations, threhsold for convergence, probability of deflecting + going in the intended directions
    self.map_grid_size = map_grid_size
    self.discount = discount

    total_states = get_total_states.prune_and_get_total_states(grid_size = self.map_grid_size)
    self.labeled_states = get_terminal_states.get_terminal_states(map_size = self.map_grid_size, total_states = total_states)
    # print([lst[0] for lst in labeled_states if np.array_equal(lst[1], np.array([1, 1, -1, 1, -1, -1, -1, 1, -1]))][0])
    # print([lst[0] for lst in labeled_states if np.array_equal(lst[1], np.array([-1, 1, 1, 1, -1, -1, -1, 1, -1]))][0])
    self.total_states, self.terminal_states = get_terminal_states.final_prune(total_states = total_states, labeled_states = self.labeled_states)
    # print([lst[0] for lst in labeled_states if np.array_equal(lst[1], np.array([1, 1, -1, 1, -1, -1, -1, 1, -1]))][0])
    # print([lst[0] for lst in labeled_states if np.array_equal(lst[1], np.array([-1, 1, 1, 1, -1, -1, -1, 1, -1]))][0])

    # Some additional parameters like total states, goal state.
    self.goal_states = [element for element in self.terminal_states if element[0] == 'Win']

    # We intialize value vector as zeros and create a copy of it which will eventually become the optimal value vector after value iteration
    self.value_vector = self.initialize_value_vector()
    self.init_val = self.value_vector

    # We intialize reward vector with 1 for win, -1 for lose and 0 for rest
    self.reward_vector = self.initialize_reward_vector()

    # We intialize the state graph. More on it in the function 'initialize_state_graph'
    self.state_paths = self.initialize_state_graph()

    # # Get Baseline policy for PolicyIteration
    # self.baseline_policy = self.initialize_baseline_policy()
    # self.optimal_policy = self.baseline_policy

    # self.optimal_value_vector, self.optimal_q_vector, self.optimal_policy = self.valueIteration() # Get optimal value, q, and policy using value iteration function - see below.
    # self.optimal_value_vector, self.optimal_q_vector, self.optimal_policy = self.policyIteration()

  def initialize_value_vector(self):
    value_vector = np.zeros(len(self.total_states))
    return value_vector

  def get_only_terminal_without_result(self):
     return [lst[1] for lst in self.terminal_states]

  def get_result_for_state(self, state, array):
      idx = [lst[0] for lst in array if np.array_equal(lst[1], np.array(state))]
      return idx[0]

  def initialize_reward_vector(self):
    reward_vector = [
    (1, lst) if result == 'Win' else
    (-1, lst) if result == 'Lose' else
    (0, lst) for result, lst in self.terminal_states
    ]
    reward_vector = reward_vector + [(0, lst) for lst in self.total_states if lst not in self.get_only_terminal_without_result()]
    return reward_vector

  def get_evaluation_of_state(self, state, player):
    terminate = False
    reward_for_state = self.get_result_for_state(state = state, array = self.reward_vector)
    if reward_for_state == 1 or reward_for_state == -1:
      terminate = True
    elif reward_for_state == 0:
      # Check if its ongoing for now then terminate false else terminate true
      result = self.get_result_for_state(state = state, array = self.labeled_states)
      if result == 'Draw':
        terminate = True
      else:
        terminate = False
    # Reward opposing for the two agents
    if player == 1:
      reward_for_state = -1 * reward_for_state
    # print(f'state: {state}, reward: {reward_for_state}, terminate: {terminate}')
    return reward_for_state, terminate

  # This function gives us the transition probabilities to each possible state from each state corresponding to each action.
  def get_transition_states(self, state, action, player):

    transitions = []
    new_state = state.copy()
    # print(f'action: {action}')
    new_state[action] = player
    reward_for_state, terminate = self.get_evaluation_of_state(state = new_state, player = player)

    return [(player, action, new_state, reward_for_state, terminate)]

  def get_all_possible_actions(self, state):
      return [index for index, value in enumerate(state) if value == 0]

  def get_states_after_player_1_move(self, state_graph):
    return [element[2] for lst in state_graph for element in lst]

  def eval_which_player_to_play(self, state):
    if state.count(-1) == state.count(1):
      # print('here -1')
      return -1
    else:
      # print('here 1')
      return 1

  # The key idea of this function is to iterate over all the states, for all actions get all possible transitions from it with their probability and reward for that transitioned state
  # The function used to get these transition probabilites, their rewards is the function 'get_transition_controlled_probabilities'. See that for more details.
  def initialize_state_graph(self):
    state_paths = {}
    for state in self.total_states:
      if state not in self.get_only_terminal_without_result():
        # print(f'state outside function call: {state}')
        actions = self.get_all_possible_actions(state = state)

        some_player_state_graph = [self.get_transition_states(state, action, self.eval_which_player_to_play(state)) for action in actions]
        # print(f'some_player_state_graph: {some_player_state_graph}')
        updated_states_after_players_move = self.get_states_after_player_1_move(state_graph = some_player_state_graph)

        # print(f'updated_states: {updated_states_after_players_move}')
        other_players_state_graph = [ 
        self.get_transition_states(state2, action, self.eval_which_player_to_play(state2))
        for state2 in updated_states_after_players_move
        if state2 not in self.get_only_terminal_without_result()
        for action in self.get_all_possible_actions(state2)
        ]

        state_paths[tuple(state)] = some_player_state_graph + other_players_state_graph
    return state_paths

  # Value iteration Function
  def valueIteration(self):

    for i in range(self.iter):
      # For each iteration, we loop over the states and for each action compute the expected future discount summed reward take the minimum of it and update the value of that state.
      delta = 0
      for state in range(self.total_states):
          if state in self.terminal_states: # i.e. if terminal state keep the value at 0
            continue
          else:
            v = self.value_vector[state]
            q_values = np.zeros(len(self.actions)) # q_values represent the all possible expected future discount summed reward for each action
            for action in list(self.actions.values()):
                for next_state, prob, reward, _ in self.state_graph[state][action]:
                    q_values[action] += prob * (reward + self.discount * self.value_vector[next_state])
            # One found the q-values for all the action update the value of the state with the maximum one and check for convergence
            self.value_vector[state] = np.max(q_values)
            delta = max(delta, np.abs(v - self.value_vector[state]))
      if delta < self.thres:
          print(f'Value-iteration converged at iteration: {i}.')
          break

      # Once found the optimal value table/vector, find optimal policy by taking argmax row-wise of the q-table - similar process as above but this time np.argmax
      policy = np.zeros(self.total_states)
      q_vector = np.zeros((self.total_states, len(self.actions)))
      for state in range(self.total_states):
          q_values = np.zeros(len(self.actions))
          for action in list(self.actions.values()):
              for next_state, prob, reward, _ in self.state_graph[state][action]:
                  q_values[action] += prob * (reward + self.discount * self.value_vector[next_state])
          q_vector[state, :] = q_values
          policy[state] = np.argmax(q_values)

    return self.value_vector, q_vector, policy

  def initialize_baseline_policy(self):
    return np.random.randint(0, 3, self.total_states)

  def policyEvaluation(self):
    delta = 0
    for state in range(self.total_states):
        if state in self.terminal_states: # i.e. if terminal state keep the value at 0
          continue
        else:
          v = self.value_vector[state]
          action_by_current_policy = self.optimal_policy[state]
          self.value_vector[state] = np.sum([prob * (reward + self.discount * self.value_vector[next_state]) for next_state, prob, reward, _ in self.state_graph[state][action_by_current_policy]])
          delta = np.max(delta, int(np.abs(v - self.value_vector[state])))
    return delta

  def policyImprovement(self):
    policy_stable = True
    for state in range(self.total_states):
      old_action = self.optimal_policy[state]
      q_values = np.zeros(len(self.actions))
      for action in list(self.actions.values()):
          for next_state, prob, reward, _ in self.state_graph[state][action]:
              q_values[action] += prob * (reward + self.discount * self.value_vector[next_state])
      self.optimal_policy[state] = np.argmax(q_values)
      if old_action != self.optimal_policy[state]:
        policy_stable = False

    return policy_stable

  def policyIteration(self):
    for i in range(self.iter):
      while True:
        delta = 0
        delta = self.policyEvaluation()
        if delta < self.thres:
          break
      policy_stable = self.policyImprovement()

      q_vector = np.zeros((self.total_states, len(self.actions)))
      if policy_stable: # Get q_matrix
        for state in range(self.total_states):
          q_values = np.zeros(len(self.actions))
          for action in list(self.actions.values()):
              for next_state, prob, reward, _ in self.state_graph[state][action]:
                  q_values[action] += prob * (reward + self.discount * self.value_vector[next_state])
          q_vector[state, :] = q_values

    return  self.value_vector, q_vector, self.optimal_policy


```
